# b. Feature Extraction : 

## 1. Extract Storage Capacity:

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Sample data (assuming data is already loaded)
# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'], format='%B %d,%Y', errors='coerce')

# Extract year and month from 'date'
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

# Extract winning team
data['winning_team'] = data['match_winner']

# Ensure margin is numerical
# First, handle any missing values
data['margin'] = data['margin'].replace('', np.nan)
data['margin'] = pd.to_numeric(data['margin'], errors='coerce')

# Sample output to verify extraction
print(data[['date', 'year', 'month', 'winning_team', 'margin']].head())
```
